```python
import pandas as pd
from ts2vg import NaturalVG
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

```python
db=pd.read_csv("D:\\python tasks\\Electric_Production.csv")
```

```python

time_series=db["IPG2211A2N"]
# type(time_series)
time_series=time_series.to_numpy()
type(time_series)
```

```python
vg=NaturalVG()
vg.build(time_series)
```

```python
edges=vg.edges
edges.sort()
```

```python
def print_edges(edges):
        for i in range(len(edges)):
                print(edges[i][0]+1,end=",")
                print(edges[i][1]+1)
```

```python
# print_edges(edges)
nxg = vg.as_networkx()
nx.draw_networkx(nxg, pos=vg.node_positions())
```

```python
time=db["DATE"]
fig1, ax1 = plt.subplots()
ax1.plot(time,time_series)
```

```python
pos = nx.spring_layout(nxg) 
nx.draw_networkx(nxg, ax=ax1, pos=pos)
ax1.plot(time_series)
plt.show()
plt.savefig(fname="save.png")
```
